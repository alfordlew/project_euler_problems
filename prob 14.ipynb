```python
def collatz(n):
    sequence = n
    count = 1
    collatz_dict = {}
    while sequence != 1:
        
        if sequence%2 == 0:
            sequence = sequence/2
        else:
            sequence = 3*sequence + 1
        collatz_dict[sequence] = count
        count += 1
    for z in collatz_dict:
        collatz_dict[z] = count - collatz_dict[z]
    collatz_dict[n] = count
    return count, collatz_dict
```

```python
def longest_chain(x):
    starting_dict = {}
    collatz_dict = {}
    for i in range(x,0,-1):
        try:
            starting_dict[i] = collatz_dict[i]
        except:
            collatz_results = collatz(i)
            collatz_dict.update(collatz_results[1])
            starting_dict[i] = collatz_results[0]
    return max(starting_dict.items(), key=operator.itemgetter(1))[0]
```
